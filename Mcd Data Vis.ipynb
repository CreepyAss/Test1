##### Problem 1
* How many calories does the average McDonald's value meal contain?

```python
df_calories = df[['Item','Calories']]
df_calories_sort = df_calories.sort_values('Calories', ascending=False)

df_calories_sort.head()
```

```python
cal = df.Calories
plt.figure(figsize=[15,7])
cal.plot(color='green', lw=2.5)

plt.xlabel('Item index')
plt.ylabel('Calories')
plt.ylim(0,2200)
plt.grid(True, axis='y')
plt.legend(loc=0)

# ---- text coordinate for annotate
text_x = df_calories_sort.head(1).index.values
text_y = df_calories_sort.head(1).Calories.values
plt.annotate(
    text = "Chicken McNuggets (40 piece)\n1880 Calories",
    xy = [text_x + 1, text_y],
    xytext = [text_x + 20, text_y + 100],
    arrowprops=(dict(facecolor='black'))
)

plt.show()
```

#### Problem 2
* How much do beverages, like soda or coffee, contribute to the overall caloric intake?

```python
df['Category'].unique()
```

```python
df_soda = df.query("Category == 'Beverages'")[['Category','Item','Calories']]

# df_soda.Item.unique()

soda = [
    'Coca-Cola Classic (Small)', 'Coca-Cola Classic (Medium)',
    'Coca-Cola Classic (Large)', 'Coca-Cola Classic (Child)',
    'Diet Coke (Small)', 'Diet Coke (Medium)', 'Diet Coke (Large)',
    'Diet Coke (Child)', 'Dr Pepper (Small)', 'Dr Pepper (Medium)',
    'Dr Pepper (Large)', 'Dr Pepper (Child)', 'Diet Dr Pepper (Small)',
    'Diet Dr Pepper (Medium)', 'Diet Dr Pepper (Large)',
    'Diet Dr Pepper (Child)', 'Sprite (Small)', 'Sprite (Medium)',
    'Sprite (Large)', 'Sprite (Child)'
]

df_soda.query(f"Item == {soda}", inplace=True)

soda_cal = df_soda['Calories'].sum()
```

```python
df_coffee = df.query("Category == 'Coffee & Tea'")[['Category','Item','Calories']]

# df_coffee.Item.unique()

df_coffee = df_coffee[~df_coffee['Item'].str.contains('tea|Tea')]

coffee_cal = df_coffee['Calories'].sum()
```

```python
# ----beverages (soda or coffee caloric intake)

total_cal = df['Calories'].sum()

(soda_cal + coffee_cal) / total_cal * 100
```

#### Problem 3
* Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value?

```python
# ---- select necesary columns

for i,j in enumerate(df.columns):
    print(i,j)      #[1,3,5,7,10,12,14,16,18,19]
```

```python
nutrition_col = [1,3,5,7,10,12,14,16,18,19]

df_sandwich = df[df['Item'].str.contains("Sandwich")].iloc[:,nutrition_col]
# df_sandwich.head()

df_grilled = df_sandwich[df_sandwich['Item'].str.contains('Grilled')]
# df_grilled

df_crispy = df_sandwich[df_sandwich['Item'].str.contains('Crispy')]
# df_crispy
```

```python
df_sandwich.columns
```

```python
# ---- barplot grilled vs. crispy
num_col = [
    'Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol',
    'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein'
]

sns.set_style('whitegrid')
plt.figure(figsize=[15,12])

for num,col in enumerate(num_col):
    df_plot = pd.DataFrame(
        data = ({
            'Grilled':[df_grilled[col].mean()],
            'Crispy':[df_crispy[col].mean()]
        })
    )
    plt.subplot(3,4,num+1)
    sns.barplot(data=df_plot)
    plt.title(col)
    plt.xticks(rotation=0)
plt.show()
```

#### Problem 4
* What about ordering egg whites instead of whole eggs?

```python
df_egg = df[df['Item'].str.contains('egg|Egg')].iloc[:,nutrition_col]
# df_egg

df_white = df_egg[df_egg['Item'].str.contains('White')]
# df_white

df_all = df_egg[~df_egg['Item'].str.contains('White')]
# df_all
```

```python
# ---- barplot white egg vs. whole egg
num_col = [
    'Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol',
    'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein'
]

sns.set_style('whitegrid')
plt.figure(figsize=[15,12])

for num,col in enumerate(num_col):
    df_plot = pd.DataFrame(
        data = ({
            'White_egg':[df_white[col].mean()],
            'Whole_egg':[df_all[col].mean()]
        })
    )
    plt.subplot(3,4,num+1)
    sns.barplot(data=df_plot)
    plt.title(col)
    plt.xticks(rotation=0)
plt.show()
```
